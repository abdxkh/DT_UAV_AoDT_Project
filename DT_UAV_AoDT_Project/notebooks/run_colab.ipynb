## 3. Plot results

```python

import pandas as pd
import matplotlib.pyplot as plt

worker = pd.read_csv('results_small/worker_train_log.csv')
manager = pd.read_csv('results_small/manager_train_log.csv')
eval_df = pd.read_csv('results_small/evaluation.csv')

display(eval_df.groupby('policy')[['avg_aodt','p95_aodt','max_aodt','mean_backhaul_energy','feasible']].mean())

plt.figure(figsize=(7,4))
plt.plot(worker['update'], worker['avg_aodt'], marker='o')
plt.xlabel('Worker PPO update')
plt.ylabel('Average AoDT')
plt.title('Worker training: AoDT')
plt.grid(True)
plt.show()

plt.figure(figsize=(7,4))
plt.plot(manager['update'], manager['avg_aodt'], marker='o')
plt.xlabel('Manager PPO update')
plt.ylabel('Average AoDT')
plt.title('Manager training: AoDT')
plt.grid(True)
plt.show()

plt.figure(figsize=(7,4))
plt.plot(manager['update'], manager['max_Z'], marker='o')
plt.xlabel('Manager PPO update')
plt.ylabel('Max virtual queue')
plt.title('Lyapunov queue pressure')
plt.grid(True)
plt.show()
```
